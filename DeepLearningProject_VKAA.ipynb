```python
# Imputing the missing values in numerical columns with the mean
for col in ['passenger_count', 'RatecodeID', 'congestion_surcharge']:
    march[col] = march[col].fillna(march[col].mean())

# Imputing missing values in the 'store_and_fwd_flag' column with the mode
march['store_and_fwd_flag'] = march['store_and_fwd_flag'].fillna(march['store_and_fwd_flag'].mode()[0])

march.isnull().sum()
```

###As there were no airports functioning during the month of march due to covid there is no data available for the airport_fee column and will hence eliminate it.


```python
march = march.drop(['airport_fee'], axis=1)
```

```python
# Identifying the outliers
numerical_cols = march.select_dtypes(include=np.number).columns
Q1 = march[numerical_cols].quantile(0.25)
Q3 = march[numerical_cols].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_identified = march[((march[numerical_cols] < lower_bound) | (march[numerical_cols] > upper_bound)).any(axis=1)]
print(f"Number of outliers identified: {len(outliers_identified)}")
```

```python
# Capping the outliers in numerical columns
for col in numerical_cols:
    march[col] = np.where(march[col] < lower_bound[col], lower_bound[col], march[col])
    march[col] = np.where(march[col] > upper_bound[col], upper_bound[col], march[col])

print("Outliers capped successfully.")
```

```python
outliers_identified = march[((march[numerical_cols] < lower_bound) | (march[numerical_cols] > upper_bound)).any(axis=1)]
print(f"Number of outliers identified: {len(outliers_identified)}")
```

```python
# Converting 'store_and_fwd_flag' to numerical using one-hot encoding
march = pd.get_dummies(march, columns=['store_and_fwd_flag'], prefix='store_and_fwd')

# Extracting numerical features from datetime columns
march['pickup_hour'] = march['tpep_pickup_datetime'].dt.hour
march['pickup_minute'] = march['tpep_pickup_datetime'].dt.minute
march['pickup_day_of_week'] = march['tpep_pickup_datetime'].dt.dayofweek
march['pickup_month'] = march['tpep_pickup_datetime'].dt.month
march['pickup_year'] = march['tpep_pickup_datetime'].dt.year

march['dropoff_hour'] = march['tpep_dropoff_datetime'].dt.hour
march['dropoff_minute'] = march['tpep_dropoff_datetime'].dt.minute
march['dropoff_day_of_week'] = march['tpep_dropoff_datetime'].dt.dayofweek
march['dropoff_month'] = march['tpep_dropoff_datetime'].dt.month
march['dropoff_year'] = march['tpep_dropoff_datetime'].dt.year

# Dropping original datetime columns
march = march.drop(['tpep_pickup_datetime', 'tpep_dropoff_datetime'], axis=1)

display(march.head())
```

```python
# Calculate the correlation matrix
correlation_matrix = march.corr(numeric_only=True)

# Display the correlation matrix
print("Correlation Matrix:")
display(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```

##From the heatmap, we can observe:

High positive correlation between fare_amount and total_amount: This is expected as total_amount is likely a sum of fare_amount and other charges. This high correlation led us to drop total_amount during dimensionality reduction to avoid multicollinearity.
High positive correlation between trip_distance and fare_amount: This is also expected, as longer trips generally result in higher fares.
High positive correlation between pickup_hour and dropoff_hour and pickup_day_of_week and dropoff_day_of_week: This is logical as pickup and dropoff times and days are closely related for a single trip.
High negative correlation between extra and VendorID: This suggests that different vendors might have different pricing structures for extra charges.
Moderate positive correlation between tip_amount and fare_amount: This indicates that customers tend to tip more on higher fares.
Low correlations between many other features: This is good for modeling as it suggests that these features provide unique information and are not redundant.
Based on these observations, we can infer some relationships within the data. For instance, trip distance is a strong driver of the fare amount, and tipping behavior is influenced by the fare. The differences in 'extra' charges between vendors could be an area for further investigation.

```python
may.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

march['fare_per_mile'] = march.apply(lambda row: row['fare_amount'] / row['trip_distance'] if row['trip_distance'] > 0 else 0, axis=1)
may['fare_per_mile'] = may.apply(lambda row: row['fare_amount'] / row['trip_distance'] if row['trip_distance'] > 0 else 0, axis=1)

X_train = march[['trip_distance', 'fare_amount', 'total_amount']]
y_train = march['fare_per_mile']

X_test = may[['trip_distance', 'fare_amount', 'total_amount']]
y_test = may['fare_per_mile']

X_train, X_test = X_train.align(X_test, join='inner', axis=1)
model = DecisionTreeRegressor(max_depth=5, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error on May data: {mse:.2f}")
print(f"R-squared on May data: {r2:.2f}")

feature_importance = model.feature_importances_
feature_importance_series = pd.Series(feature_importance, index=X_train.columns)
sorted_feature_importance = feature_importance_series.sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x=sorted_feature_importance.values, y=sorted_feature_importance.index, palette='viridis')
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title('Feature Importance from Decision Tree Model (March data, evaluated on May data)')
plt.show()
```

```python
#Setting a threshold for high correlation
threshold = 0.9

#Creating a mask to select only the upper triangle of the correlation matrix (excluding the diagonal)
upper = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))

#For features that are highly correlated (absolute correlation > threshold)
to_drop = [column for column in upper.columns if any(upper[column].abs() > threshold)]

print(f"Features to drop due to high correlation: {to_drop}")

#Dropping the highly correlated features
january = january.drop(columns=to_drop)

print("\nDataFrame after dimensionality reduction:")
display(january.head())
```

## Conclusion of EDA

###Findings from the Comparison

*   Total taxi trip volume in March 2020 was over 53% lower than in January 2020, decreasing from over 6.4 million trips to just over 3 million trips.
*   The distributions of trip distance and passenger count showed a reduction in the frequency of trips across all ranges in March compared to January.
*   The average tip amount slightly decreased in March compared to January.
*   Despite the significant drop in overall volume, the relative proportions of payment types and vendor usage remained largely consistent between the two months.
*   Hourly and daily taxi trip volumes in March showed a sharp decline throughout the month, particularly in the latter half, with less pronounced peak hours compared to January.

###Steps for future

*   The dramatic decrease in taxi ridership in March 2020 is a clear indicator of the immediate and significant impact of the early stages of the COVID-19 pandemic and associated public health measures on urban mobility in NYC.
*   Further analysis could explore if the types of trips being taken changed (e.g., essential travel vs. leisure) and how this might have influenced metrics like trip distance, passenger count, and tips per trip among the remaining trips.
